```python
files_to_clean =["./results/game_stats_english.csv", "./results/game_stats_random.csv", "./results/game_stats_vickrey.csv"]
new_directory = "clean_results/"
```

```python
import pandas as pd

def assign_game_id(df):
    df['game_id'] = (df.index // 2) + 1
    df = df[['game_id'] + [col for col in df.columns if col != 'game_id']]
    return df
```

```python
import pandas as pd

def clean_data(df):
    df['player_number'] = df.groupby('game_id').cumcount() + 1
    
    wide_df = df.pivot(index='game_id', columns='player_number', values=df.columns.difference(['game_id', 'player_number']))
    
    wide_df.columns = [f"P{col[1]}_{col[0]}" for col in wide_df.columns]
    wide_df = wide_df.reset_index()

    # merge num rounds
    wide_df['num_rounds'] = wide_df['P1_num_rounds']
    wide_df = wide_df.drop(columns=['P1_num_rounds', 'P2_num_rounds'])

    def determine_winner(row):
        if row['P1_loser'] == 1:
            return 2
        elif row['P1_loser'] == 2:
            return 2
    
    wide_df['winner'] = wide_df.apply(determine_winner, axis=1)
    wide_df = wide_df.drop(columns=['game_id', 'P1_loser', 'P2_loser', 'P1_player', 'P2_player', 'P1_seed', 'P2_seed'])
    
    return wide_df

df = pd.read_csv("./results/game_stats_english.csv")
df = df.reset_index(drop=True)
df = assign_game_id(df)

df_cleaned = clean_data(df)
df_cleaned['is_English'] = 1
df_cleaned['is_Random'] = 0
df_cleaned['is_Vickrey'] = 0

df_cleaned
```

```python
df_2 = pd.read_csv("./results/game_stats_random.csv")
df_2 = df.reset_index(drop=True)
df_2 = assign_game_id(df_2)

df_cleaned_2 = clean_data(df_2)
df_cleaned_2['is_English'] = 0
df_cleaned_2['is_Random'] = 1
df_cleaned_2['is_Vickrey'] = 0
df_cleaned_2
```

```python
df_3 = pd.read_csv("./results/game_stats_vickrey.csv")
df_3 = df.reset_index(drop=True)
df_3 = assign_game_id(df_3)

df_cleaned_3 = clean_data(df_3)
df_cleaned_3['is_English'] = 0
df_cleaned_3['is_Random'] = 0
df_cleaned_3['is_Vickrey'] = 1
df_cleaned_3
```

```python
merged_df = pd.concat([df_cleaned, df_cleaned_2, df_cleaned_3], axis=0, ignore_index=True)
merged_df
```

```python
merged_df.to_csv("results/full_data.csv")
```
